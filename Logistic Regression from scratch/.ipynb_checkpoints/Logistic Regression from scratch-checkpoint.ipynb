```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def data_cleaning():
    
    fx = open("hw6x.dat", "r")
    data = fx.readlines()
    fy = open("hw6y.dat", "r")
    label = fy.readlines() 
    X_, y = [], []

    for line in data:
        values = line.split()
        x, y_ = float(values[0]), float(values[1])
        X_.append([x, y_])
    X_ = np.asarray(X_)

    [y.append(float(line)) for line in label]

    b = np.ones((X_.shape[0], 1)) # intercept
    X = np.concatenate((X_,b), axis=1) # adding intercept form
    
    return([X,y])

def sigmoid(X, theta):
    
    z = np.dot(X, theta)
    return (1/(1 + np.exp(-z)))
    
def gradient(X, y, sigmoid_z):
    
     return (np.dot(X.T, (sigmoid_z - y)) / len(y))

def gradient_descent(X, y):
    
    theta = np.zeros(X.shape[1]) # weights initialized
    start = time.time()
    for i in range(0,50000):
        
        sigmoid_z = sigmoid(X, theta)
        dw_gradient = gradient(X, y, sigmoid_z)

        theta = theta - (0.01 * dw_gradient)   # considering step size = 0.01
        
    total_time = (time.time() - start)
    return ([theta, total_time])

def newton_method(X, y):
    
    theta = np.zeros(X.shape[1]) # weights initialized
    start = time.time()
    
    for i in range(0,1000):
        sigmoid_z = sigmoid(X, theta)
        dw_gradient = gradient(X, y, sigmoid_z)

        R_ = (sigmoid_z * (1-sigmoid_z)) # f''(x) = X.T H X where H is diagonal matrix of sig(z)*(1-sig(z))
        R = np.diag(np.diag(R_))
        R = np.reshape(R, (99,1))
        c = np.multiply(X,R)
        c2 = np.dot(c.T, X)
        c3 = np.linalg.inv(c2)         # f'/f'' = dot product of f' and inv(f'')

        c4 = np.dot(c3,dw_gradient)

        theta = theta - c4             # no step size is used
    total_time = (time.time() - start)
    return ([theta, total_time])

def loss(X, y, theta):
    
    sigmoid_z = sigmoid(X, theta)
    ones = np.ones(X.shape[0])
    l = ((y * np.log(sigmoid_z) + (ones - y) * np.log(1 - sigmoid_z)).mean())
    return (round(l,5))

def plot(X, y):
    
    X1p, X2p, X1n, X2n, X1 = [], [], [], [], []
    # data with prob > 0.5
    for i in range(0, len(y)):    # plot 1s and 0s for x1 and x2 as axes. Separating out x1, x2 for 1s and 0s 
        X1.append(X[i][0])
        if (y[i] == 1.0):
            X1p.append(X[i][0]) 
            X2p.append(X[i][1])
        else:
            X1n.append(X[i][0]) 
            X2n.append(X[i][1])

    # getting the x co-ordinates of the decision boundary
    plot_x = np.array([min(X[:,0]) - 2, max(X[:,0]) + 2])
    # getting corresponding y co-ordinates of the decision boundary
    plot_y = (-1/theta[1]) * (theta[0] * plot_x + theta[2])

    plt.figure(num=None, figsize=(8, 8), dpi=100, facecolor='w', edgecolor='k')
    plt.plot(X1p, X2p, 'bo', marker='+', label="Positive/Ones points")
    plt.plot(X1n, X2n, 'ro', marker='^', label="Negative/Zeros points")
    plt.plot(plot_x, plot_y, 'k', label = "Decision_Boundary")
    plt.legend(loc='best')
    plt.title("Logistic Regression")
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.show()

def AUC_score(X, theta):
    p_, n_ = 49, 50
    sigmoid_z = sigmoid(X, theta)
    deno = (p_ * n_)
    c = 0
    for i in range(0, p_):   
        for j in range(0, n_):
            if (sigmoid_z[j] <= sigmoid_z[(i+50)]):   # Checking of 0s prob is <= 1s prob, then add 1
                c = c + 1            
    return (c / deno)


if __name__ == "__main__":
    
    # clean data
    data = data_cleaning() 
    
    # Get Theta Values for each method
    theta_gradient_descent, time1 = gradient_descent(data[0], data[1]) 
    theta_newton_method, time2 = newton_method(data[0], data[1])
    
    # Generate plot and calculate AUC score
    plot(data[0], data[1])
    AUC = AUC_score(data[0], theta_newton_method) 
    
    # Calculate Loss from each method
    loss_gradient_descent = loss(data[0], data[1], theta_gradient_descent) 
    loss_newton_method = loss(data[0], data[1], theta_newton_method)
    
    # Generate Outputs
    print("Method \t\t Algorithm Run time \t\t #Iterations \t\t Coefficients \t\t\t Loss ")
    print("-------\t\t -------------------\t\t ------------\t\t -------------\t\t\t -----")
    print(f"Gradient_desc \t {time1} \t\t 50,000 \t {theta_gradient_descent} \t {loss_gradient_descent}")
    print(f"Newton Method \t {time2} \t\t 01,000 \t {theta_newton_method} \t {loss_newton_method}")

    print(f"\nAUC socre is: {AUC}")
```

Observations:

1) Gradient Descent uses much more number of iterations to arrive at the same precision of theta values. <br>
2) Algorithm run time is a lot more in Gradient Descent.

Conclusion:

Newton Method is much faster and requires less iterations and no step size. <br>

Note: Used iteration increment method to check the #iterations required in gradient descent method to match theta precision.
