# Linear Regression from scratch

```python
# ** Amith K Singh  **
# Import Libraries
import matplotlib.pyplot as plt
import numpy as np
from numpy import linalg as LA

# Accessing file and converting to numeric
file = open("hw2.dat","r");
data = file.readlines()
x, y, sum_ = [], [], 0

for line in data:
    values = line.split()
    x.append(float(values[0])), y.append(float(values[1]))

for a in range(1,8):    # Loop over to generate all the plots in single run
    if (a%2 != 0):
        
        print("\n********** LLSE of degree %s **********" %a)
        N = a+1               # For 1,3,5,7
        X = np.vander(x,N)    # Vandermonde Matrix

        # Computation part to find the coef matrix using 'Normal Equation'
        X_prod = np.matmul(X.T,X)
        inverse = np.linalg.inv(X_prod)
        Y_prod = np.matmul(X.T,y)

        coef_matrix = np.matmul(inverse, Y_prod)
        print("Coefficent or Parameters are: \n" + str(coef_matrix))

        y_pred = np.matmul(X, coef_matrix)
        
        # Calculate Errors - MSE, RMSE and Absolute Error (L2)
        mse=np.mean((y-y_pred)**2)
        print("MSE is %s " %mse)

        rmse = np.sqrt(np.mean((y-y_pred)**2))
        print("RMSE is %s " %rmse)

        print ("Absolute error (L2-Norm): %s" %LA.norm(y-y_pred))
        
        # Plot the result
        plt.figure(num=None, figsize=(7, 7), dpi=100, facecolor='w', edgecolor='k')
        plt.plot(x, y, 'bo', label="Actual data")
        plt.plot(x, np.matmul(X, coef_matrix), 'r-', label=("Best fit: deg=" + str(N-1)))
        plt.legend(loc='best')
        plt.title("Lineat Least Squares - Polynomial with deg-%s" %a)
        plt.xlabel("x - Data points")
        plt.ylabel("y - predicted/Actual Value")
        plt.show()
```
